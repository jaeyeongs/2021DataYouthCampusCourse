### 베르누이 나이브베이즈
- 데이터 : https://www.kaggle.com/team-ai/spam-text-message-classification

```python
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("spam.csv")
data.head()
```

```python
data.info()
```

```python
data["label"] = data["Category"].map({"spam":1, "ham":0})
data.head()
```

```python
X = data["Message"]
y = data["label"]

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 103)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
cv = CountVectorizer(max_features=1000, binary=True) #binary : 이진모델(0 과 1)
x_train_cv = cv.fit_transform(x_train)
```

```python
encoded = x_train_cv.toarray()
encoded
```

```python
cv.inverse_transform(encoded[0]) # 0번째 메세지에 아래와 같은 단어들이 있다는 뜻
```

```python
cv.get_feature_names()
```

```python
len(cv.get_feature_names())
```

#### 베르누이 나이브베이즈 분류

```python
nb_clf = BernoulliNB()
nb_clf.fit(x_train_cv, y_train)
```

```python
x_test_cv = cv.fit_transform(x_test)
x_test_cv
```

```python
print(x_test_cv)
```

```python
encoded2 = x_test_cv.toarray()
encoded2
```

```python
pred = nb_clf.predict(x_test_cv)
```

```python
accuracy_score(y_test, pred)
```
