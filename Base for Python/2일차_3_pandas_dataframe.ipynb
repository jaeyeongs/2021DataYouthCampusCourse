### <연습문제>
- #### Age의 NaN 값을 다음과 같이 처리하시오.
    - 생존자(Survived = 1)는 생존자 나이의 평균으로 대체
    - 사망자(Survived = 0)는 사망자 나이의 평균으로 대체

```python
# 생존자 나이 평균
sur1_mean = train[train['Survived'] == 1]['Age'].mean() # train 데이터의 생존자(Survived)가 1인 train 데이터의 나이(Age) 평균

# 사망자 나이 평균
sur0_mean = train[train['Survived'] == 0]['Age'].mean() # train 데이터의 사망자(Survived)가 0인 train 데이터의 나이(Age) 평균

print(sur1_mean, sur0_mean)
```

```python
# train 데이터의 생존자(Survived)가 1인 train 데이터의 나이(Age) 결측값에 평균 채우기
train[train['Survived'] == 1]['Age'].fillna(sur1_mean)
```

```python
# train 데이터의 생존자(Survived)가 0인 train 데이터의 나이(Age) 결측값에 평균 채우기
train[train['Survived'] == 0]['Age'].fillna(sur0_mean)
```

```python
train # 실제 train 데이터에 적용되지 않음
```

```python
# train 데이터 NaN 값에 실제로 데이터 채우기
train.loc[train["Survived"] == 1, "Age"] = train[train["Survived"] == 1]["Age"].fillna(sur1_mean)
```

```python
train.info() # Age 변수에 생존자 평균 나이를 766개 까지 채움
```

```python
train.loc[train["Survived"] == 0, "Age"] = train[train["Survived"] == 0]["Age"].fillna(sur0_mean)
```

```python
train.info() # Age 변수에 사망자 평균 나이를 891개 까지 채워서 NaN 값 없이 전부 채움
```

- #### merge() 함수

```python
fruits = pd.DataFrame({'customer_id' : [1, 3, 2, 5, 4, 3, 5, 8],
                      "fruit" : ["apple", "banana", "lemon", "melon", "peach", "banana", "peach", "lemon"],
                      "num" : [5, 3, 10, 1, 3, 2, 7, 5]})
fruits
```

```python
customer = pd.DataFrame({'customer_id' : np.arange(5),
                      "name" : ["Pengsoo", "Pororo", "Ryan", "Apeach", "Neo"]})
customer
```

```python
pd.merge(customer, fruits, on = "customer_id", how = "inner") # inner가 디폴트 값
```

```python
pd.merge(customer, fruits, on = "customer_id", how = "left")
```

```python
pd.merge(customer, fruits, on = "customer_id", how = "right")
```

```python
pd.merge(customer, fruits, on = "customer_id", how = "outer")
```

```python
cust_id = customer.set_index("customer_id")
fruits_id = fruits.set_index("customer_id")
```

```python
cust_id
```

```python
fruits_id
```

```python
pd.merge(cust_id, fruits_id, left_index = True, right_index = True)
```

- (예제1) 가장 많이 팔린 과일은?

```python
pd.merge(customer, fruits, on = "customer_id").groupby("fruit").sum().sort_values(by = "num", ascending = False)
```

- (예제2) Apeach가 가장 많이 구매한 과일은?

```python
pd.merge(customer, fruits, on = "customer_id").groupby(["name","fruit"]).sum().loc["Apeach"]
```
