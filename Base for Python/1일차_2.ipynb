### 반복문_while

- #### while

```python
a = [1, 3, 9, 5]

i = 0 #인덱스
while i < len(a):
    print("index: ", i, "value: ",  a[i])
    i = i + 1
```

```python
a = [1, 3, 9, 5]

i = 0 #인덱스
while i < len(a):
    if a[i] > 5:
        print("index: ", i, "value: ",  a[i])
    i = i + 1
```

- #### 무한루프

```python
while True:
    print("end")
```

- #### break

```python
a = [1, 3, 9, 5]

i = 0 #인덱스
while i < len(a):
    if a[i] > 5:
        break
    print("index: ", i, "value: ",  a[i])
    i = i + 1
```

### 반복문_for

- #### for

```python
a = [1, 3, 9, 5]

for i in a:
    print(i)
    
print("end")
```

```python
a = [1, 3, 9, 5]

for i in a:
    print(i, i * 2)
    
print("end")
```

```python
a = [1, 3, 9, 5]

for index, num in enumerate(a):
    print("index: ", index, num)
    
print("end")
```

```python
a = {"A": "apple", "B":"banana", "C":"carrot"}

for key, value in a.items():
    print(key, value)
```

- #### break

```python
a = [1, 3, 9, 5]
for num in a:
    if num > 5:
        break
        
    print(num)
        
print("end")
```

- #### 중첩

```python
a = [1, 3, 9, 5]
for i in a:
    for j in a:
         print(i ,"*",j, "=", i*j)
print("end")
```

### <연습문제>

- #### 구구단을 2 - 9단까지 출력하세요. 

```python
# 풀이1 - for 이용
a = [2, 3, 4, 5, 6, 7, 8, 9]
b = list(range(1, 10))

for i in a:
    for j in b:
        print(i ,"*",j, "=", i*j)

print("end")
```

```python
# 풀이2 - while 이용
x = 2
while x <= 9:
    y = 1
    while y <= 9:
        print(x, "x", y, "=", x*y)
        y = y + 1 
    x = x + 1
```

- #### 1 - 100 까지 정수 중 3의 배수 또는 7의 배수를 모두 출력하세요.

```python
# 내 풀이
a = list(range(1, 101))
i = 1
while i <= 101:
  if i % 3 == 0 or i % 7 == 0:
    print(i)
  i = i + 1
```

```python
# 교수님 풀이
b = list(range(1, 101))

for i in b:
    if i % 3 == 0 or i % 7 == 0:
        print(i)
```

- #### a = [3, 25, 34, 87, 55, 49, 7, 97, 63, 28] 에서 최대값, 최소값, 평균을 구하세요. 

```python
# 내 풀이
a = [3, 25, 34, 87, 55, 49, 7, 97, 63, 28]

print("최대값 :", max(a)) # 최대값
print("최소값 :", min(a)) # 최소값
print("평균 :", sum(a) / len(a)) # 평균
```

```python
# 교수님 풀이1
b = [3, 25, 34, 87, 55, 49, 7, 97, 63, 28]
b.sort() # 작은 숫자부터 정렬
print("최소값 :", b[0]) # b리스트의 제일 첫번째 값
print("최대값 :", b[-1]) # b리스트의 제일 마지막 값
```

```python
# 교수님 풀이2
# 최소값
b = [3, 25, 34, 87, 55, 49, 7, 97, 63, 28]

_min = b[0]
for x in b:
    if x < _min:
        _min = x
print(_min)
```

```python
# 최대값
b = [3, 25, 34, 87, 55, 49, 7, 97, 63, 28]

_max = b[0]
for x in b:
    if x > _max:
        _max = x
print(_max)
```

```python
# 최소값 & 최대값 동시 출력
b = [3, 25, 34, 87, 55, 49, 7, 97, 63, 28]

_min = b[0]
_max = b[0]
for x in b:
    if x < _min:
        _min = x
    if x > _max:
        _max = x
        
print(_min, _max)
```

```python
# 평균
b = [3, 25, 34, 87, 55, 49, 7, 97, 63, 28]

i = 0
_sum = 0
while i < len(b):
    _sum = _sum + b[i]
    i = i + 1
print(_sum)
print(len(b))
print(_sum / len(b))
```

```python
# 교수님 풀이3
b = [3, 25, 34, 87, 55, 49, 7, 97, 63, 28]

_sum = 0
for i in b:
    _sum = _sum + i

print(_sum / len(b))
```
