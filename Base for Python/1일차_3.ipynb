### <연습문제>

- #### 주어진 숫자 리스트의 평균을 구하는 함수를 정의하시오.

```python
# 내 풀이
def mean(nums):
    return sum(nums) / len(nums) 
```

```python
a = [1, 2, 3, 4, 5]
print(mean(a))
```

```python
# 교수님 풀이1
def mean(lists):
    return sum(lists) / len(lists)
```

```python
print(mean([1, 2, 3, 4, 5]))
```

```python
# 교수님 풀이2
b = [1, 2, 3, 4, 5]
print(mean(b))
```

- #### 이름과 나이를 입력받아 자기소개하는 함수를 정의하시오.

```python
# 내 풀이
def say_hello(name, msg):
    print("안녕하세요, 저는", name, msg)
```

```python
name = "신재영입니다."
msg = "반갑습니다."
say_hello(name, msg)
```

```python
# 교수님 풀이
def say_myself(name, age):
    print("나의 이름은 {}입니다.".format(name))
    print("나이는 {}세 입니다.".format(age))
```

```python
print(say_myself("신재영", 26))
```

- #### 두개의 숫자를 입력받아 사칙연산 결과를 출력하는 함수를 정의하시오.

```python
def calc(x, y):
    add = x + y
    sub = x - y
    div = x / y
    mult = x * y
    return add, sub, div, mult
```

```python
calc(1, 2)
```
